```python
# Define known variables
density = 1000   # kg/m^3
volume = 0.5     # m^3

# Calculate mass
mass = density * volume
print("Mass of the fluid:", mass, "kg")
```

```python
# Simple plot: Pressure vs Volume for Ideal Gas (P = nRT/V)
import numpy as np
import matplotlib.pyplot as plt

V = np.linspace(0.1, 1.0, 100)
n = 1.0       # moles
R = 8.314     # J/molÂ·K
T = 300       # K
P = n * R * T / V

plt.plot(V, P)
plt.xlabel("Volume (mÂ³)")
plt.ylabel("Pressure (Pa)")
plt.title("Pressure vs Volume (Ideal Gas)")
plt.grid(True)
plt.show()
```
